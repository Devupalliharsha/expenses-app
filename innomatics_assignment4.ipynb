Problem 1: Mobile Recharge Validation System


```python
def start_recharge():
    # List of allowed recharge plans
    approved_plans = [199, 299, 399, 599]

    active = True
    while active:
        user_choice = input("Enter recharge amount (or 'exit' to stop): ")

        if user_choice.lower() == 'exit':
            break

        # Change input to number
        money = int(user_choice)

        # Check if it meets the rules
        if money >= 50 and money in approved_plans:
            print("Recharge Successful! Processing...")
            active = False # This stops the loop
        else:
            print("That is not a valid plan. Please try again.")

# Call the function
start_recharge()
```

Problem 2: Inventory Reorder Alert System


```python
def check_stock(items_dict):
    for product in items_dict:
        quantity = items_dict[product]

        # If quantity is under 15, we need more
        if quantity < 15:
            print("Low Stock Alert:", product, "only has", quantity, "left.")
        else:
            print(product, "is fine. Count:", quantity)

# Dictionary of store items
my_inventory = {"Chargers": 10, "Cases": 25, "ScreenGuards": 5, "Earphones": 30}
check_stock(my_inventory)
```

 Problem 3: Student Result Processing System


```python
def find_result(marks_list):
    sum_total = 0
    # Add up every score in the list
    for score in marks_list:
        sum_total = sum_total + score

    # Calculate the average
    final_avg = sum_total / len(marks_list)

    print("Average Score:", final_avg)

    if final_avg >= 50:
        return "Result: Pass"
    else:
        return "Result: Fail"

# Test it out
my_scores = [45, 60, 55, 30]
print(find_result(my_scores))
```

Problem 4: Cab Fare Estimator with Retry Option


```python
def calculate_cab_fare(distance, is_rush_hour):
    base_fee = 50
    per_km = 12
    total = base_fee + (distance * per_km)

    # Add 25% extra if it's peak time
    if is_rush_hour == "yes":
        total = total + (total * 0.25)

    return total

while True:
    km = float(input("How many kilometers? "))
    peak = input("Is it peak hour? (yes/no): ").lower()

    price = calculate_cab_fare(km, peak)
    print("Your total fare is:", price)

    retry = input("Do you want to check another ride? (y/n): ")
    if retry.lower() != 'y':
        print("Thank you!")
        break
```

Problem 5: Employee Attendance Eligibility Checker


```python
def check_my_attendance(days_list):
    count_p = 0
    total_days = len(days_list)

    # Loop to count how many 'P' are there
    for day in days_list:
        if day == "P":
            count_p += 1

    percentage = (count_p / total_days) * 100
    print("Attendance Percentage:", percentage, "%")

    if percentage >= 75:
        return "Status: Eligible"
    else:
        return "Status: Not Eligible"

#im using a Sample attendance record
intern_log = ["P", "P", "A", "P", "P", "P", "A", "P"]
print(check_my_attendance(intern_log))
```

Problem 6: Password Strength Checker


```python
def is_password_safe(code):
    # Check length first
    if len(code) < 8:
        return "Too Short! Must be 8+ characters."

    has_num = False
    has_sym = False
    symbols = ["@", "#", "$"]

    # Check every letter in the password
    for char in code:
        if char.isdigit():
            has_num = True
        if char in symbols:
            has_sym = True

    # Check if both requirements were met
    if has_num == True and has_sym == True:
        return "Strong Password"
    else:
        return "Weak Password: Add a number or @#$"

# Testing
test_pw = input("Enter a password to test: ")
print(is_password_safe(test_pw))
```
